# statistical tests

```python
from micom.stats import compare_groups

long_exports = merged[merged.direction == "export"]
long_exports["flux"] = long_exports["flux"].abs()
```

long_imports comes from merged, which comes from Prevotella, meaning it contains the exchange fluxes

```python
long_exports
```

```python
tests = compare_groups(long_exports, "medium")
```

```python
tests.sort_values(by="q")
```

```python
from plotnine import *
theme_set(theme_minimal())

selected_mets = tests.sort_values(by="q").metabolite[0:4] 

pl = (
    ggplot(long_exports[long_exports.metabolite.isin(selected_mets)])
    + aes(x="medium", y="flux")
    + geom_boxplot() 
    + geom_jitter(width=.2)
    + scale_y_log10()
    + facet_wrap("~ metabolite")
)
pl
```
