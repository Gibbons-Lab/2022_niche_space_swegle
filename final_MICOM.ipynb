Importing the three mediums that will be used for MICOM

```python
from micom.qiime_formats import load_qiime_medium
```

```python
uploaded_fiber = load_qiime_medium("/proj/gibbons/refs/micom_media/media/vmh_high_fiber_agora.qza")
uploaded_western = load_qiime_medium("/proj/gibbons/refs/micom_media/media/western_diet_gut_agora.qza")
uploaded_fermented = load_qiime_medium("/proj/gibbons/refs/micom_media/media/vmh_fermented_agora.qza")
```

```python
uploaded_western["medium"] = "western"
uploaded_fiber["medium"] = "high_fiber"
uploaded_fermented["medium"] = "fermented"
merged = pd.concat([uploaded_western, uploaded_fiber, uploaded_fermented])
merged
```

```python
per_sample_medium = pd.merge(pre_merged, merged, on="medium")
```
