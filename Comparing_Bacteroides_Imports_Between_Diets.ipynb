next step 
- comparing the metabolite imports of the most abundant taxa before and after diet intervention

```python
western_only = taxon_imports_table.loc[taxon_imports_table['medium'] == 'western']
western_only
```

```python
western_bacteroides_only = western_only.loc[western_only['taxon'] == 'Bacteroides']
western_bacteroides_only
```

```python
mean_import_of_western_bacteroides = western_bacteroides_only["flux"].mean()
mean_import_of_western_bacteroides
```

```python
mean_import_of_western_bacteroides = western_bacteroides_only.groupby('name')["flux"].mean()
mean_import_of_western_bacteroides
```

```python
high_fiber_only = taxon_imports_table.loc[taxon_imports_table['medium'] == 'high_fiber']
high_fiber_only
```

```python
high_fiber_bacteroides_only = high_fiber_only.loc[high_fiber_only['taxon'] == 'Bacteroides']
high_fiber_bacteroides_only
```

```python
mean_import_of_high_fiber_bacteroides = high_fiber_bacteroides_only["flux"].mean()
mean_import_of_high_fiber_bacteroides
```

```python
mean_import_of_high_fiber_bacteroides = high_fiber_bacteroides_only.groupby('name')["flux"].mean()
mean_import_of_high_fiber_bacteroides
```

```python
mean_import_of_western_bacteroides
```

```python
mean_import_of_high_fiber_bacteroides
```

```python
western_bacteroides_imports = pd.DataFrame (mean_import_of_western_bacteroides, columns = ['flux'])
```

```python
western_bacteroides_imports.rename(columns={"flux": "western_flux"}, inplace=True)
```

```python
western_bacteroides_imports
```

```python
high_fiber_bacteroides_imports = pd.DataFrame (mean_import_of_high_fiber_bacteroides, columns = ['flux'])
```

```python
high_fiber_bacteroides_imports.rename(columns={"flux": "high_fiber_flux"}, inplace=True)
```

```python
high_fiber_bacteroides_imports
```

```python
combined_diet_fluxes = pd.concat([western_bacteroides_imports['western_flux'], high_fiber_bacteroides_imports['high_fiber_flux']], axis=1, keys= ['western_bacteroides_imports', 'high_fiber_bacteroides_imports'])
```

```python
combined_diet_fluxes
```

```python
bacteroides_imports_for_both_diets = combined_diet_fluxes.dropna()
```

```python
bacteroides_imports_for_both_diets
```

next step
- subtract the mean imports between the two diets
- see which metabolites have the largest switches between diets 

```python
bacteroides_difference_between_means = bacteroides_imports_for_both_diets["western_bacteroides_imports"]/bacteroides_imports_for_both_diets["high_fiber_bacteroides_imports"]
```

```python
bacteroides_difference_between_means
```

```python
bacteroides_imports_for_both_diets['Bacteroides_Difference_Between_Means'] = bacteroides_difference_between_means
```

```python
bacteroides_imports_for_both_diets
```

next step
- finding the metabolites that have the biggest absolute value of means 

```python
bacteroides_imports_for_both_diets['Bacteroides_Difference_Between_Means'] = bacteroides_imports_for_both_diets['Bacteroides_Difference_Between_Means'].abs()
bacteroides_imports_for_both_diets
```

```python
bacteroides_imports_for_both_diets.nlargest(10, ['Bacteroides_Difference_Between_Means'])
```
